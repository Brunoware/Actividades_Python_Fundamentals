# Alumnos con menor puntuación de lectura
1. Buscar el min en "reading_score"
2. Guardar el dataframe como reading_min **Tip: Elegir todas las columnas**
2. Crear gender_scname con columnas gender y school name
2. Obtener los dummies por género y guardar en dummies_genero
3. Hacer un merge (concat) entre estos dos dataframes y llamarlo "plotbygen"
4. Obtener el head
5. Agrupar utilizando la columna school_name por medio de una suma y guardar en "plot_grouped"
6. Crear un gráfico de columnas

```python
# paso 1 y 2
reading_min=data_students.loc[data_students.reading_score==data_students.reading_score.min()].copy()
# paso 3
gender_scname=reading_min[['gender','school_name']]
# paso 4 y 5
dummies_genero=pd.concat([gender_scname.gender,pd.get_dummies(gender_scname,columns=['gender'])],axis=1)
# paso 6
dummies_genero.head()
```

```python
# paso 7
plot_grouped=dummies_genero.groupby(['school_name']).sum()

# paso 8
plot_grouped.plot(kind='bar',figsize=(12,6),rot=45)
plt.legend(['F','M'])
plt.show()
```

# Definir una función con dos entradas
1. La función se debe llamar Func1 y se debe graficar los alumnos por escuela y genero con x puntaje de y asignatura **Tip:Cuidado con los cambios de variable y nombres de dataframes** Func1(reading_score,63) Func1(asignatura,puntaje) Func1(y,x)

```python
def func1(asignatura,puntaje):
    # paso 1 y 2
    reading_min=data_students.loc[data_students[asignatura]==puntaje].copy()
    # paso 3
    gender_scname=reading_min[['gender','school_name']]
    # paso 4 y 5
    dummies_genero=pd.concat([gender_scname.gender,pd.get_dummies(gender_scname,columns=['gender'])],axis=1)
    # paso 7
    plot_grouped=dummies_genero.groupby(['school_name']).sum()
    # paso 8
    plot_grouped.plot(kind='bar',figsize=(12,6),rot=45)
    plt.legend(['F','M'])
    plt.show()
```

```python
func1('math_score',79)
```
