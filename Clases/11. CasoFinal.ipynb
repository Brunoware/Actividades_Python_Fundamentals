# Corte por género
1.-Crear gender_scname con columnas gender y school name

2.-Obtener los dummies por género y guardar en dummies_genero

3.-Hacer un merge (concat) entre estos dos dataframes y llamarlo "plotbygen"

4.-Obtener el head

5.-Agrupar utilizando la columna school name por medio de una suma y guardar en "plot_grouped"

6.-Crear un gráfico de columnas

```python
gender_scname=data_students[["gender","school_name"]] #1
dummies_genero=pd.get_dummies(gender_scname[["gender"]]) #2
plotbygen=pd.concat([gender_scname,dummies_genero], axis=1) #3
plotbygen.head() #4
```

```python
plot_grouped=plotbygen.groupby(["school_name"]).sum() #5
```

```python
plot_grouped.plot(kind="bar", color=["springgreen","chocolate"], figsize=(10,4)) #6
```

```python
reading_min=data_students[data_students["reading_score"]==data_students["reading_score"].min()]
gender_scname=reading_min[["gender","school_name"]]
dummies_genero=pd.get_dummies(gender_scname[["gender"]])
plotbygen=pd.concat([gender_scname,dummies_genero], axis=1)
plot_grouped=plotbygen.groupby(["school_name"]).sum()
plot_grouped.plot(kind="bar", color=["springgreen","chocolate"], figsize=(10,4)) #6
```

```python
def Func1(y,x):
    df1=data_students[x==y]
    gender_scname=df1[["gender","school_name"]]
    dummies_genero=pd.get_dummies(gender_scname[["gender"]])
    plotbygen=pd.concat([gender_scname,dummies_genero], axis=1)
    plot_grouped=plotbygen.groupby(["school_name"]).sum()
    print("Gáfico con puntaje de ", x)
    return plot_grouped.plot(kind="bar", color=["springgreen","chocolate"], figsize=(10,4))
```

```python
def Func2(y,x):
    print("Gráfico de ", y, "con puntaje de ", x)
    y=data_students[y]
    df1=data_students[x==y]
    gender_scname=df1[["gender","school_name"]]
    dummies_genero=pd.get_dummies(gender_scname[["gender"]])
    plotbygen=pd.concat([gender_scname,dummies_genero], axis=1)
    plot_grouped=plotbygen.groupby(["school_name"]).sum()
    return plot_grouped.plot(kind="bar", color=["springgreen","chocolate"], figsize=(10,4))
```
