# Importar librerías y cargar datos

Importar librerías

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

Cargar datos

```python
df=pd.read_csv('datasets/athlete_events.csv')
```

# Actividad: Básica 

Crear un programa en Visual Studio que me permita saber cuál es el competidor más veterano que ha recibido medalla (oro, plata o bronce)

```python
# se escoge una muestra limpia del dataframe para la variables 'Medal'
sample=df.loc[(df.Age.notna())&(df.Medal.notna())]
# se buscar al competidor mayor con alguna medalla
competidor_veterano=sample.loc[sample.Age==sample.Age.max()]
# se muestra el registro del competidor
competidor_veterano
```

Crear un programa en Visual Studio que me permita saber cuál es el competidor más joven que ha recibido medalla de oro

```python
# se extrae un dataframe de solo medallas de oro
df_oro=df.loc[df.Medal=='Gold']
# se buscar al competidor más joven (pueden ser más)
competidor_joven_oro=df_oro.loc[df_oro.Age==df_oro.Age.min()]
# se muestran los datos
competidor_joven_oro
```

Encuentra al competidor más ganador de la historia y crea un csv con toda su información.

```python
# se eliminan los registros duplicados en una muestra para no contar varias veces un registro; sin embargo, no se limpia los nan
# pues se considera que si existe un registro de una competencia no repetida es porque ganó alguna medalla pese a no estar
# registrada
df_ganador=df.drop_duplicates(keep='first')
# .value_counts() ordena descendetemente por defecto,por lo que el mayor es el nombre del primer registro
nombre_ganador_historico=df_ganador.Name.value_counts().index[0]
# buscar datos y mostrarlos
datos_ganador=df_ganador.loc[df_ganador.Name==nombre_ganador_historico]
# exportar datos del ganador
datos_ganador.to_csv('datasets/datos_ganador_historico_actividad_basica.csv',index=False)
```

```python
# mostrar datos del ganador histórico (comprabando que se pueda leer del .csv), se aprecia que son registros únicos
pd.read_csv('datasets/datos_ganador_historico_actividad_basica.csv').head()
```

# Actividad: Avanzada

Crear un programa en Visual Studio que me permita saber cuál es el competidor más veterano que ha recibido medalla para los NOC´s MEX, COL y ARG (oro, plata o bronce) 

```python
# como se busca obtener la medalla, se limpian los valores nan de esta
df_noc=df.loc[df.Medal.notna()]
# se realiza el filtro para los NOC's propuestos
df_noc=df_noc.loc[df_noc.NOC.isin(['MEX','COL','ARG'])]
# Se eliminan duplicados por para ['Name','NOC'] para quedarnos con nombres únicos en todos los NOC's; esto sirve
# para extraerlos posteriormente
df_noc_no_duplicates=df_noc.drop_duplicates(subset=['Name','NOC'],keep='first')
# se entiende que debe existir un grupo de personas que han recibido medalla para los tres NOC's a la vez (según enunciado)
# ; y, de este grupo, se debe buscar al mayor, entonces se usa .value_counts() y se trae los nombres que aparecen tres veces
# (en caso existan)
df_count_names=pd.DataFrame(df_noc_no_duplicates.Name.value_counts()).reset_index().rename(
                            columns={'index':'Name','Name':'Count_of_names'})
# se busca los nombres cuyo 'Counts_of_name' sea de 3 (MEX,COL,ARG)
names=df_count_names.loc[df_count_names.Count_of_names==3,'Name'].values
names
```

```python
# se observa que el array names esta vacío, por lo que se entrega el más veterano del dataframe 'df_noc'
df_noc.loc[df_noc.Age==df_noc.Age.max()]
```

Crear un programa en Visual Studio que me permita saber cuál es el competidor más joven que ha recibido medalla de oro para los NOC´s USA y CAN

```python
# como se busca obtener la medalla, se limpian los valores nan de esta
df_noc=df.loc[df.Medal.notna()]
# se realiza el filtro para los NOC's propuestos
df_noc=df_noc.loc[df_noc.NOC.isin(['USA','CAN'])]
# Se eliminan duplicados por para ['Name','NOC'] para quedarnos con nombres únicos en todos los NOC's; esto sirve
# para extraerlos posteriormente
df_noc_no_duplicates=df_noc.drop_duplicates(subset=['Name','NOC'],keep='first')
# se entiende que debe existir un grupo de personas que han recibido medalla para los dos NOC's a la vez (según enunciado)
# ; y, de este grupo, se debe buscar al menor, entonces se usa .value_counts() y se trae los nombres que aparecen dos veces
# (en caso existan)
df_count_names=pd.DataFrame(df_noc_no_duplicates.Name.value_counts()).reset_index().rename(
                            columns={'index':'Name','Name':'Count_of_names'})
# se busca los nombres cuyo 'Counts_of_name' sea de 2 (USA,CAN)
names=df_count_names.loc[df_count_names.Count_of_names==2,'Name'].values
names
```

```python
# se observa que el array names esta vacío, por lo que se entrega el más joven del dataframe 'df_noc'
df_noc.loc[df_noc.Age==df_noc.Age.min()]
```

Encuentra al competidor más ganador de la historia en medallas totales, medallas de oro, medallas de plata y medallas de broce para los NOC´s USA, CHN y RUS. Crea un csv con toda su información.

```python
# a partir de los dos ejercicios anteriores, ya no se busca un grupo de personas en los todos los grupos NOC de este caso 
# (USA,CHN,RUS) con el fin de evitar un array 'name' vacío, por lo que se realiza primero es elimnar registros duplicados 
# como tal, y , posteriormente contar sus apariciones en el subset para los NOC's seleccionados, no se limpia las medallas 
# nan pues se comprende que si esta en el registro es porque gano alguna de estas; sin embargo no fue anotado y el registró 
# quedó nulo
df_ganador=df.drop_duplicates(keep='first')
# se realiza el filtro para los NOC's propuestos
df_noc=df_ganador.loc[df_ganador.NOC.isin(['USA','CHN','RUS'])]
# .value_counts() ordena descendetemente por defecto,por lo que el mayor es el nombre del primer registro
nombre_ganador_historico=df_noc.Name.value_counts().index[0]
# buscar datos y mostrarlos
datos_ganador=df_noc.loc[df_noc.Name==nombre_ganador_historico]
# exportar datos del ganador
datos_ganador.to_csv('datasets/datos_ganador_historico_actividad_avanzada.csv',index=False)
```
